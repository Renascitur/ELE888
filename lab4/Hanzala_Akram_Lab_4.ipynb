```python
from pandas.core.arrays.sparse import array
import matplotlib
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
%matplotlib inline 
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Functions #
# Point graphing function
def plotPoints(xPoints,yPoints,title,xLabel,yLabel):
  plot = plt.scatter(xPoints,yPoints)
  plt.title(title)
  plt.xlabel(xLabel)
  plt.ylabel(yLabel)
  return plot

# Create the initial values based on equation given and save them in a dataframe so when we shuffle x values allign with corresponding y values
def createValuesFunc(amount):
  x = np.linspace(-1, 1, num = amount)
  y = (0.2*x**4) + (2*x**3) + (0.1*x**2) + 10
  df = pd.DataFrame(x, columns = ["x"])
  df["y"] = y
  return df

# Shuffle values if argument is true
def shuffleData(argument, data):
  if (argument == True):
    data = shuffle(data)
  return data

# Split data into three parts
def splitData(data, trainRatio, validationRatio, testRatio):
  trainAndValData, testData = train_test_split(data, test_size=testRatio)
  valRatio = validationRatio/(1-testRatio)
  trainData, validationData = train_test_split(trainAndValData, test_size=valRatio)
  return trainData, validationData, testData

# Scale data between 0 and 1
def scaleData(data):
  scaler = MinMaxScaler()
  scaledData = scaler.fit_transform(data)
  scaledData = pd.DataFrame(scaledData, columns=["x", "y"])
  return scaledData

# Calculate different types of errors
def errorMetricFunc(yTest,yPredicted):
  MAE = mean_absolute_error(yTest, yPredicted)
  MSE = mean_squared_error(yTest, yPredicted, squared=True)
  RMSE = mean_squared_error(yTest, yPredicted, squared=False)
  R2Score = r2_score(yTest, yPredicted)
  return MAE, MSE, RMSE, R2Score

# Create a neural netowork using structure one model
def structureOne(train, validation, test, activationFunction):
  # Create layers
  structOne = Sequential()
  structOne.add(Dense(12, activation=activationFunction, input_dim=1))
  structOne.add(Dense(8, activation=activationFunction))
  structOne.add(Dense(4, activation=activationFunction))
  structOne.add(Dense(1, activation=activationFunction))
  # Compile function using values asked for
  structOne.compile(loss="mse", optimizer="adam", metrics=["mse"])
  structOne.fit(train["x"], train["y"], epochs=20, batch_size=12, validation_data=(validation["x"], validation["y"]))
  structOne.summary()
  yPredicted = structOne.predict(test["x"])
  return yPredicted

# Create a neural netowork using structure two model
def structureTwo(train, validation, test, activationFunction):
  # Create layers
  structTwo = Sequential()
  structTwo.add(Dense(24, activation=activationFunction, input_dim=1))
  structTwo.add(Dense(1, activation=activationFunction))
  # Compile function using values asked for
  structTwo.compile(loss="mse", optimizer="adam", metrics=["mse"])
  structTwo.fit(train["x"], train["y"], epochs=20, batch_size=12, validation_data=(validation["x"], validation["y"]))
  structTwo.summary()
  yPredicted = structTwo.predict(test["x"])
  return yPredicted

# ---------- Q1 ---------- #
# Create data with 30 000 points from given function
data = createValuesFunc(30000)
# Plot data
initialGraph = plotPoints(data["x"], data["y"], "Initial Graph", "x", "y")
plt.show()
print("\n")

# ---------- Q2 ---------- #
# Shuffle the data if True
shuffledData = shuffleData(True, data)
shuffledGraph = plotPoints(shuffledData["x"], shuffledData["y"], "Shuffled Graph", "x", "y")
plt.show()
print("\n")

# ---------- Q3 ---------- #
# Split data into different sets with length depending on ratios
train, validation, test = splitData(data, 0.3, 0.2, 0.5)
print("Data split into:")
print(f"Training data shape: {train.shape}")
print(f"Validation data shape: {validation.shape}")
print(f"Test data shape: {test.shape}\n")

# ---------- Q4 ---------- #
# Normalize data between 0 and 1
scaledData = scaleData(data)
scaleddGraph = plotPoints(scaledData["x"], scaledData["y"], "Scaled Graph", "x", "y")
plt.show()
print("\n")

# ---------- Q5 ---------- #
# Find values for different scoring metrics
# MAE, MSE, RMSE, R2 = errorMetricFunc(yPredicted, test["y"])

# ---------- Case 1.1: Shuffled, Structure 1, relu ---------- #
caseData = shuffledData
train, validation, test = splitData(caseData, 0.3, 0.2, 0.5)
yPredicted = structureOne(train, validation, test, "relu")
MAE, MSE, RMSE, R2 = errorMetricFunc(test["y"], yPredicted)
print("\nMAE is:", MAE)
print("MSE is:", MSE)
print("RMSE is:", RMSE)
print("R2 Score is:", R2, "\n")

caseTest = plotPoints(test["x"], test["y"], "Case 1.1: xTest vs yTest", "x", "y")
plt.show()
print("\n")
casePredict = plotPoints(test["x"], yPredicted, "Case 1.1: xTest vs yPredicted", "x", "y")
plt.show()
print("\n")

# ---------- Case 1.2: Shuffled, Structure 2, relu ---------- #
caseData = shuffledData
train, validation, test = splitData(caseData, 0.3, 0.2, 0.5)
yPredicted = structureTwo(train, validation, test, "relu")
MAE, MSE, RMSE, R2 = errorMetricFunc(test["y"], yPredicted)
print("\nMAE is:", MAE)
print("MSE is:", MSE)
print("RMSE is:", RMSE)
print("R2 Score is:", R2, "\n")

caseTest = plotPoints(test["x"], test["y"], "Case 1.2: xTest vs yTest", "x", "y")
plt.show()
print("\n")
casePredict = plotPoints(test["x"], yPredicted, "Case 1.2: xTest vs yPredicted", "x", "y")
plt.show()
print("\n")

# ---------- Case 1.3: Shuffled, Structure 1, tanh ---------- #
caseData = shuffledData
train, validation, test = splitData(caseData, 0.3, 0.2, 0.5)
yPredicted = structureOne(train, validation, test, "tanh")
MAE, MSE, RMSE, R2 = errorMetricFunc(test["y"], yPredicted)
print("\nMAE is:", MAE)
print("MSE is:", MSE)
print("RMSE is:", RMSE)
print("R2 Score is:", R2, "\n")

caseTest = plotPoints(test["x"], test["y"], "Case 1.3: xTest vs yTest", "x", "y")
plt.show()
print("\n")
casePredict = plotPoints(test["x"], yPredicted, "Case 1.3: xTest vs yPredicted", "x", "y")
plt.show()
print("\n")

# ---------- Case 1.4: Shuffled, Scaled, Structure 1, relu ---------- #
caseData = shuffledData
train, validation, test = splitData(caseData, 0.3, 0.2, 0.5)
train = scaleData(train)
validation = scaleData(validation)
test = scaleData(test)
yPredicted = structureOne(train, validation, test, "relu")
MAE, MSE, RMSE, R2 = errorMetricFunc(test["y"], yPredicted)
print("\nMAE is:", MAE)
print("MSE is:", MSE)
print("RMSE is:", RMSE)
print("R2 Score is:", R2, "\n")

caseTest = plotPoints(test["x"], test["y"], "Case 1.4: xTest vs yTest", "x", "y")
plt.show()
print("\n")
casePredict = plotPoints(test["x"], yPredicted, "Case 1.4: xTest vs yPredicted", "x", "y")
plt.show()
print("\n")

# ---------- Case 1.5: Shuffled, Scaled, Structure 1, tanh ---------- #
caseData = shuffledData
train, validation, test = splitData(caseData, 0.3, 0.2, 0.5)
train = scaleData(train)
validation = scaleData(validation)
test = scaleData(test)
yPredicted = structureOne(train, validation, test, "tanh")
MAE, MSE, RMSE, R2 = errorMetricFunc(test["y"], yPredicted)
print("\nMAE is:", MAE)
print("MSE is:", MSE)
print("RMSE is:", RMSE)
print("R2 Score is:", R2, "\n")

caseTest = plotPoints(test["x"], test["y"], "Case 1.5: xTest vs yTest", "x", "y")
plt.show()
print("\n")
casePredict = plotPoints(test["x"], yPredicted, "Case 1.5: xTest vs yPredicted", "x", "y")
plt.show()
print("\n")

# ---------- Case 2.1: Unshuffled, Structure 1, relu ---------- #
caseData = data
train, validation, test = splitData(caseData, 0.3, 0.2, 0.5)
yPredicted = structureOne(train, validation, test, "relu")
MAE, MSE, RMSE, R2 = errorMetricFunc(test["y"], yPredicted)
print("\nMAE is:", MAE)
print("MSE is:", MSE)
print("RMSE is:", RMSE)
print("R2 Score is:", R2, "\n")

caseTest = plotPoints(test["x"], test["y"], "Case 2.1: xTest vs yTest", "x", "y")
plt.show()
print("\n")
casePredict = plotPoints(test["x"], yPredicted, "Case 2.1: xTest vs yPredicted", "x", "y")
plt.show()
print("\n")

# ---------- Case 2.2: Unshuffled, Structure 2, relu ---------- #
caseData = data
train, validation, test = splitData(caseData, 0.3, 0.2, 0.5)
yPredicted = structureTwo(train, validation, test, "relu")
MAE, MSE, RMSE, R2 = errorMetricFunc(test["y"], yPredicted)
print("\nMAE is:", MAE)
print("MSE is:", MSE)
print("RMSE is:", RMSE)
print("R2 Score is:", R2, "\n")

caseTest = plotPoints(test["x"], test["y"], "Case 2.2: xTest vs yTest", "x", "y")
plt.show()
print("\n")
casePredict = plotPoints(test["x"], yPredicted, "Case 2.2: xTest vs yPredicted", "x", "y")
plt.show()
print("\n")

# ---------- Case 2.3: Unshuffled, Structure 1, tanh ---------- #
caseData = data
train, validation, test = splitData(caseData, 0.3, 0.2, 0.5)
yPredicted = structureOne(train, validation, test, "tanh")
MAE, MSE, RMSE, R2 = errorMetricFunc(test["y"], yPredicted)
print("\nMAE is:", MAE)
print("MSE is:", MSE)
print("RMSE is:", RMSE)
print("R2 Score is:", R2, "\n")

caseTest = plotPoints(test["x"], test["y"], "Case 2.3: xTest vs yTest", "x", "y")
plt.show()
print("\n")
casePredict = plotPoints(test["x"], yPredicted, "Case 2.3: xTest vs yPredicted", "x", "y")
plt.show()
print("\n")

# ---------- Case 2.4: Unshuffled, Scaled, Structure 1, relu ---------- #
caseData = data
train, validation, test = splitData(caseData, 0.3, 0.2, 0.5)
train = scaleData(train)
validation = scaleData(validation)
test = scaleData(test)
yPredicted = structureOne(train, validation, test, "relu")
MAE, MSE, RMSE, R2 = errorMetricFunc(test["y"], yPredicted)
print("\nMAE is:", MAE)
print("MSE is:", MSE)
print("RMSE is:", RMSE)
print("R2 Score is:", R2, "\n")

caseTest = plotPoints(test["x"], test["y"], "Case 2.4: xTest vs yTest", "x", "y")
plt.show()
print("\n")
casePredict = plotPoints(test["x"], yPredicted, "Case 2.4: xTest vs yPredicted", "x", "y")
plt.show()
print("\n")

# ---------- Case 2.5: Unshuffled, Scaled, Structure 1, tanh ---------- #
caseData = data
train, validation, test = splitData(caseData, 0.3, 0.2, 0.5)
train = scaleData(train)
validation = scaleData(validation)
test = scaleData(test)
yPredicted = structureOne(train, validation, test, "tanh")
MAE, MSE, RMSE, R2 = errorMetricFunc(test["y"], yPredicted)
print("\nMAE is:", MAE)
print("MSE is:", MSE)
print("RMSE is:", RMSE)
print("R2 Score is:", R2, "\n")

caseTest = plotPoints(test["x"], test["y"], "Case 2.5: xTest vs yTest", "x", "y")
plt.show()
print("\n")
casePredict = plotPoints(test["x"], yPredicted, "Case 2.5: xTest vs yPredicted", "x", "y")
plt.show()
print("\n")
```
